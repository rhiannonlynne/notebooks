```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from astropy import units as u
from astropy.coordinates import SkyCoord
```

```python
# Set up LSST-sky ra/dec grid.
step = 20
ra = np.arange(0, 360+step/2, step)
ra = np.concatenate([ra, np.array([359.9])])
dec = np.arange(-90, 30+step/4, step/2)
r, d = np.meshgrid(ra, dec)
d2, r2 = np.meshgrid(dec, ra)
```

```python
sky = SkyCoord(ra=r*u.deg, dec=d*u.deg, frame='icrs')
sky2 = SkyCoord(ra=r2*u.deg, dec=d2*u.deg, frame='icrs')
```

```python
plt.plot(sky.ra.deg, sky.dec.deg, 'k.')
for r, d in zip(sky.ra.deg, sky.dec.deg):
    plt.plot(r, d, 'r-')
plt.xlabel('RA')
plt.ylabel('Dec')
plt.xlim(0, 360)
plt.ylim(-90, 90)
```

```python
plt.plot(sky.galactic.l.deg, sky.galactic.b.deg, 'k.')
plt.xlabel('Galactic l')
plt.ylabel('Galactic b')
```

```python
# set up onstant dec lines that don't cross image .. 
l_dec = sky.galactic.l.deg
b_dec = sky.galactic.b.deg
for l, b in zip(l_dec, b_dec):
    plt.plot(l, b, 'k.')
plt.plot(l_dec[-1], b_dec[-1], 'k-')
plt.fill(l_dec[-1], b_dec[-1])
plt.xlabel('Galactic l')
plt.ylabel('Galactic b')
plt.xlim(0, 360)
plt.ylim(-90, 90)
```
